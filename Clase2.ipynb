```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

Arreglo de numero ordenados y numeros en medio

```python
x_1= np.linspace(-5,5,100)
```

```python
x_1
```

```python
u=np.random.normal(0,1,100)
```

numeros ceudualatorios

```python
u
```

```python
y= 4 + 2 * x_1 + u
```

```python
y
```

Grafico : diagrama de dispersion

```python
plt.scatter(x_1,y)
plt.show()
```

```python
X=np.c_[np.ones(len(x_1)),x_1]
```

```python
X
```

ecuacion minimo cuadrados ordinarios

```python
betas=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
```

```python
betas
```

Variable con gorrito


```python
y_est=betas[0]+betas[1]*x_1
```

```python
y_est
```

graficar

```python

fig,ax = plt.subplots(1,1,figsize=(10,4))
ax.plot(x_1, y_est, color="r", label="model")
ax.scatter(x_1, y, color="k", label="observations")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("MCO")
ax.legend()
plt.show()
```
